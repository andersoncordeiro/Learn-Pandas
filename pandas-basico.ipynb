```python
#importando as bibliotecas para análise dos dados

import numpy as np 
import pandas as pd 
```

Vamos importar nosso conjunto de dados principal da imigração canadense usando o método read_excel() do pandas.

```python
dfCanada = pd.read_excel('Canada.xlsx', sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter=2)
```

Vamos ver as 5 primeiras linhas do conjunto de dados usando a função `head()`.

```python
# você pode especificar o número de linhas que gostaria de ver da seguinte maneira: dfCanada.head (10)
# por padrão, o comando head exibe as 5 primeiras
dfCanada.head()
```

Também podemos ver as últimas 5 linhas do conjunto de dados usando a função `tail ()`.

```python
dfCanada.tail()
```

Ao analisar um conjunto de dados, é sempre uma boa ideia começar obtendo informações básicas sobre o seu dataframe. Podemos fazer isso usando o método `info ()`.

Este método pode ser usado para obter um breve resumo do dataframe.

```python
# a opção verbose = False apresenta um resumo do dataframe, com verbose = True nós temos também 
# informações sobre cada coluna

dfCanada.info(verbose=False)
```

Para obter a lista de cabeçalhos de coluna, podemos chamar o parâmetro `.columns` do dataframe.

```python
dfCanada.columns
```

Da mesma forma, para obter a lista de índices, usamos o parâmetro `.index`.

```python
dfCanada.index.values
```

Para visualizar as dimensões do dataframe, usamos o parâmetro `.shape`.

```python
# tamanho do dataframe (linhas, colunas)
dfCanada.shape    
```

Vamos limpar o conjunto de dados para remover algumas colunas desnecessárias. 
Podemos usar o método _pandas_ `drop ()` da seguinte maneira:

```python
# em pandas o eixo = 0 representa as linhas (padrão) e o eixo = 1 representa as colunas.
dfCanada.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
dfCanada.head(2)
```

Vamos renomear as colunas para que façam sentido. Podemos usar o método `rename ()` passando um dicionário de nomes antigos e novos da seguinte maneira:

```python
dfCanada.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
dfCanada.columns
```

Também adicionaremos uma coluna 'Total' que resume o total de imigrantes por país durante todo o período de 1980 a 2013, da seguinte forma:

```python
dfCanada['Total'] = dfCanada.sum(axis=1)
```

Finalmente, vamos ver um resumo de cada coluna em nosso dataframe usando o método `describe ()`.

```python
dfCanada.describe()
```
